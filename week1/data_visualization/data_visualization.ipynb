## Bar Charts and Heatmaps

````python
flight = pd.read_csv("data_sets/flight_delays.csv",index_col="Month")
````

````python
flight
````

````python
plt.figure(figsize=(10,6))
plt.title("Barchart of flight delays")
sns.barplot(x=flight.index,y=flight['NK'])
plt.ylabel("Arrival delay (in minutes)")
````

````python
plt.figure(figsize=(14,7))
plt.title("Heatmap of flight delays")
sns.heatmap(data=flight,annot=True)
plt.xlabel("Airline")
````

````python
plt.figure(figsize=(14,7))
plt.title("Heatmap of flight delays")
sns.heatmap(data=flight)
plt.xlabel("Airline")
````

## Scatter Plots

````python
sns.scatterplot(x=insurance['bmi'],y=insurance['charges'])
````

## Distributions

- Histograms
- Kernel density estimate (KDE) plot

````python
iris = pd.read_csv('data_sets/iris.csv', index_col="Id")
iris.head()
````

````python
sns.histplot(iris['Petal Length (cm)'])
````

````python
sns.kdeplot(data=iris['Petal Length (cm)'], fill=True)
````

````python
sns.jointplot(x=iris['Petal Length (cm)'], y=iris['Sepal Width (cm)'],kind="kde")
````

````python
sns.histplot(data=iris, x='Petal Length (cm)', hue='Species')
````

````python
sns.kdeplot(data=iris, x='Petal Length (cm)', hue='Species', fill=True)
plt.title("Distribution of Petal Lengths, by Species")
````
