## Feature Selection

```python
X = train_data.drop(['failure'],axis=1)
y = train_data['failure']
X.head()
```

## Splitting Train dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
```

```python
X_train.drop(['attribute_2','attribute_3'], axis=1,inplace=True)
X_valid.drop(['attribute_2','attribute_3'], axis=1,inplace=True)
```

```python
# list of categorical variables
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
X_train.head()
```

```python
X_train_copy = X_train.copy()
X_valid_copy = X_valid.copy()
```

```python
categorical_cols = [cname for cname in X_train.columns if X_train[cname].nunique() < 10 and 
                        X_train[cname].dtype == "object"]

# Select numerical columns
numerical_cols = [cname for cname in X_train.columns if X_train[cname].dtype in ['int64', 'float64']]

# Keep selected columns only
my_cols = categorical_cols + numerical_cols
```

## Using Pipeline for data preprocessing

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=0)

from sklearn.metrics import roc_auc_score
# Bundle preprocessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', model)])

# Preprocessing of training data, fit model 
my_pipeline.fit(X_train, y_train)

# Preprocessing of validation data, get predictions
preds = my_pipeline.predict(X_valid)
print("AUC under ROC: ",roc_auc_score(y_valid, preds))
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

model1 = KNeighborsClassifier(n_neighbors=4)
pipeline1 = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model1)
                             ])
pipeline1.fit(X_train, y_train)
y_pred1 = pipeline1.predict(X_valid)
print("AUC under ROC: ",roc_auc_score(y_valid, y_pred1))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier(criterion='entropy',random_state=0)
pipeline2 = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model2)
                             ])
pipeline2.fit(X_train, y_train)
y_pred2 = pipeline2.predict(X_valid)
print("AUC under ROC: ",roc_auc_score(y_valid, y_pred2))
```

### XG Boost

```python
import xgboost as xgb

model3 = xgb.XGBClassifier(random_state=6)
pipeline3 = Pipeline(steps=[('preprocessor', preprocessor),('model', model3)])
pipeline3.fit(X_train, y_train)
y_pred3 = pipeline3.predict(X_valid)
print("AUC under ROC: ",roc_auc_score(y_valid, y_pred3))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model4 = LogisticRegression()
pipeline4 = Pipeline(steps=[('preprocessor', preprocessor),('model', model4)])
pipeline4.fit(X_train, y_train)
y_pred4 = pipeline3.predict(X_valid)
print("AUC under ROC: ",roc_auc_score(y_valid, y_pred4))
```

## Imputing and One hot encoding separately 

```python
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[object_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[object_cols]))

# One-hot encoding removed index; put it back
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

```python
OH_X_train
```

```python
from sklearn.impute import SimpleImputer

OH_X_train.columns = OH_X_train.columns.astype(str)
OH_X_valid.columns = OH_X_valid.columns.astype(str)

# Imputation
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(OH_X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(OH_X_valid))

# Imputation removed column names; put them back
imputed_X_train.columns = OH_X_train.columns
imputed_X_valid.columns = OH_X_valid.columns
imputed_X_train.index = OH_X_train.index
imputed_X_valid.index = OH_X_valid.index
```
